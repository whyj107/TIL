```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing DataSet and take a look at Data
dataset = pd.read_csv("Boston_housing.csv")
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.isnull().sum()
```

```python
data = dataset.dropna()
```

```python
data.isnull().sum()
```

```python
data.info()
data.describe()
```

```python
data.plot.scatter('RM', 'MEDV')
```

```python
# set the size of the figure
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.heatmap(data.corr().round(2), square=True, cmap='RdYlGn', annot=True)
```

```python
X = data[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y = data['MEDV']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state = 24)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
X_test
```

```python
predictions = model.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
sns.distplot((y_test-predictions),bins=50);
```

```python
new_input = [[100, 100, 2.01, 1, 0.435, 6.635, 29.7, 8.3440, 4, 280, 20.0, 390.94, 8.99]] 
```

```python
predictions = model.predict(new_input)
print(predictions)
```

# DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(max_depth = 10,random_state=24)
regressor.fit(X_train, y_train)
```

```python
X_test
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
new_input = [[2.7, 80, 2.01, 0.0, 0.435, 4.635, 29.7, 8.3440, 4, 280, 17.0, 390.94, 5.99]] 
```

```python
predictions = regressor.predict(new_input)
print(predictions)
```

R² score tells us how well our model is fitted to the data by comparing it to the average line of the dependent variable. If the score is closer to 1, then it indicates that our model performs well versus if the score is farther from 1, then it indicates that our model does not perform so well.

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
