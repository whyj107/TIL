#### 오늘의 백준 문제들은 백트래킹에 관련된 문제들이다. 백트래킹에 대해서 아무것도 모르기 때문에 문제를 풀지 않고 다른 사람들이 작성한 코드를 보면서 학습하였다.
#### 백트래킹 학습하기에 앞서 깊이 우선 탐색과 너비 우선 탐색에 대해서 학습을 하여야 했는데 카테고리를 따로 만들기가 힘들어 아래에 정리하였다.
#### 이번 주말에 오늘의 백준 문제와 백트래킹에 대해서 다시 한번 학습해야겠다.

```python
graph={
    1:[2, 3],
    2:[1, 4, 5, 7],
    3:[1, 5, 9],
    4:[2, 6],
    5:[2, 3, 7, 8],
    6:[4],
    7:[5, 2],
    8:[5],
    9:[3],
}
```

## 깊이 우선 탐색(DFS: Depth First Search)

```python
def DFS(graph, root):
    stack=[root]
    visited=[]
    while stack:
        n = stack.pop()
        for i in graph[n]:
            if i not in visited:
                stack.append(i)
        if n not in visited:
            visited.append(n)
        # print('v', visited)
    return visited
print(DFS(graph, 1))
```

## 너비 우선 탐색(BFS: Breadth First Search)

```python
def BFS(graph, root):
    visited = []
    queue = [root]
    while queue:
        n = queue.pop(0)
        if n not in visited:
            visited.append(n)
        for i in graph[n]:
            if i not in visited:
                queue.append(i)
        # print('v', visited)
    return visited

print(BFS(graph, 1))
```
