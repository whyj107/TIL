```python
import os
import numpy as np
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.models import load_model
import tensorflow.keras.backend as K
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization)
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers
```

```python
# print(tf.VERSION)
print(tf.keras.__version__)
```

```python
# Data Object
FASHION_MNIST_DATA_OBJECT = keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = FASHION_MNIST_DATA_OBJECT.load_data()
```

```python
print("Train Images Shape :", train_images.shape)
print("Test Images Shape :", test_images.shape)
print("Train Labels Shape :", train_labels.shape)
print("Test Labe;s Shape :", test_labels.shape)
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# Let us plot some training images to see how they look
plt.figure(figsize=(30,30))
for i in range(15):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

```python
train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))
```

```python
train_images[1]
```

```python
print("Train Images Shape :", train_images.shape)
print("Test Images Shape :", test_images.shape)
```

```python
# Normalization
train_images_norm = train_images / 255.0
test_images_norm = test_images / 255.0
```

```python
train_images_norm[1].shape
```

```python
#Initializing Values

IMAGE_HEIGHT, IMAGE_WIDTH, NUM_CHANNELS, NUM_CLASSES = 28, 28, 1, 10

model = Sequential()
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same', input_shape=(IMAGE_HEIGHT, IMAGE_WIDTH, NUM_CHANNELS)))
model.add(BatchNormalization())
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(256, kernel_size=(3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(256, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(10, activation='softmax'))

adam = Adam(lr=0.001, decay=1e-6)
model.compile(optimizer=adam, loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
# model.fit(train_images_norm, train_labels, epochs=3, batch_size=1024, shuffle=True, validation_split=0.3)
history = model.fit(train_images_norm, train_labels, validation_data=(test_images_norm, test_labels), epochs=5, batch_size=256 )
```

```python
test_loss, test_accuracy = model.evaluate(test_images_norm, test_labels)

print("Test Accuracy:: ", test_accuracy)
print("Test Loss ::", test_loss)
```

```python
plt.figure(figsize=(30,15))
plt.plot(model.history.history['acc'])
plt.plot(model.history.history['val_acc'])
plt.title("Accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(30,15))
plt.plot(model.history.history['loss'])
plt.plot(model.history.history['val_loss'])
plt.title("Loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# Making predictions on Test dataset
prediction_prob = model.predict(test_images_norm)
prediction_classes = np.argmax(prediction_prob, -1)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(test_labels, prediction_classes))
print(classification_report(test_labels, prediction_classes))
```
