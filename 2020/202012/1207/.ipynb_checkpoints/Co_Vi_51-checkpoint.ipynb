```python
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
from skimage.color import rgb2hsv
from skimage.transform import resize
%matplotlib inline
```

```python
img = imread('Seen.jpg', as_gray=False)
imshow(img)
```

```python
imge = imread('Seen.jpg', as_gray=True)
imshow(imge)
```

```python
print(img.shape)
```

```python
plt.subplot(121), imshow(img)
plt.title('RGB Format')

plt.subplot(122), imshow(imge)
plt.title('Grayscale Format')

plt.show()
```

```python
new_img = rgb2hsv(img)
```

```python
plt.subplot(121), imshow(img)
#plt.title()
plt.subplot(122), imshow(new_img)
```

```python
plt.subplot(121), imshow(img)
plt.title('Original Image')

img_resize = resize(img,(300,300))
plt.subplot(122), imshow(img_resize)
plt.title('Resized Image')
```

```python
from skimage.transform import rotate
image = imread('Seen.jpg')
image_rotated = rotate(image, angle=120)
imshow(image_rotated)
```

# Flip Image

```python
from numpy import fliplr, flipud
img_flip = fliplr(img)
plt.subplot(121), imshow(img)
plt.subplot(122), imshow(img_flip)
```

# Crop Image

```python
image = imread('see.jpg')

cropped = image[100:(img.shape[0]-100),100:(img.shape[1]-100)]

plt.subplot(121), imshow(image)
plt.title ('Original Image')
plt.subplot(122), imshow(cropped)
plt.title('Cropped Image')


plt.show()
```

# Altering Image Brightness 

```python
from skimage import exposure

image = imread('Seen.jpg')
image_bright = exposure.adjust_gamma(image, gamma=0.5, gain=1)
image_dark = exposure.adjust_gamma(image, gamma=1.5, gain=1)

plt.subplot(131), imshow(image)
plt.title('Original Image')

plt.subplot(132), imshow(image_bright)
plt.title('Bright Image')

plt.subplot(133), imshow(image_dark)
plt.title('Dark Image')

plt.show()
```

```python
from skimage.filters import median

image = imread('Seen.jpg', as_gray=True)
image_median = median(image)

plt.subplot(121), imshow(image)
plt.title('Original Image')

plt.subplot(122), imshow(image_median)
plt.title('Smooth Image')

plt.show()
```

# Open CV

### Image Contours

```python
# pip install opencv-python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('math.jpg')
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(gray_image,1277,255,0)
```

```python
#im, contours,hierarchy = cv2.findContours(thresh,cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# Grayscale 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
  
# Find Canny edges 
edged = cv2.Canny(gray, 30, 200) 
cv2.waitKey(0)

contours, hierarchy = cv2.findContours(edged,  
    cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
with_contours = cv2.drawContours(image,contours,-1,(0,255,0),3)
plt.imshow(with_contours)
```
