배열 내 모든 요소에 스칼라 값을 더할 수 있다.
y = x + 2

v1의 각 요소에 1을 더하고 그 결과를 변수 r에 저장
r = v1 + 1

동일한 크기를 갖는 두 배열을 서로 더할 수 있다.
z = x + y

벡터 v1 및 벡터 v2의 합인 벡터 vs를 만들어라
vs = v1 + v2

배열 내 모든 요소에 스칼라 값을 곱하거나 나눌 수 있다.
z = 2*x
y = x/3

값 vs를 2로 나눈 값을 포함하는 변수 va를 만들어라
va = vs/2

벡터의 최댓값은 max 함수를 사용하여 구할 수 있다.
xMax = max(x)

xSqrt = sqrt(x)
xRound = round(x)

* : 크기가 동일한 두 개의 벡터를 곱
.* : 요소별 곱셈

plot 함수를 사용하여 길이가 같은 두 개의 벡터를 x축, y축 좌표로 하는 그래프를 플로팅할 수 있다.
plot(x, y)

plot 함수는 색, 선 스타일, 마커 스타일을 지정하는 인수를 추가로 받는다.
plot(x, y, "r--o")
o마커와 빨간색 파선을 사용하여 플로팅

선 없고 빨간색 별마커
plot(x, y, "r  *")

이미 그려진 선 위에 다른 선을 겹쳐서 플로팅하려면 hold on을 사용하면 된다.
hold on

작업
hold on 명령을 입력하세요.
그런 다음 sample(x축)에 대해 mass1(y축)을 플로팅하세요. 이때 선은 그리지 않고 검은색(k) 정사각형 마커(s)만 표시되도록 하세요.
hold on
plot(sample, mass1, "k  s")

디폴트 동작으로 되돌려서 각 플롯마다 좌표축을 가지도록 하려면 hold off를 입력한다.
hold off

하나의 벡터만 사용하여 플로팅하면 벡터의 값들이 y축 데이터로 사용되고 x축 데이터는 1부터 n(이 벡터에 포함된 요소의 개수)으로 지정된다.
plot(v1)

plot 함수는 속성 이름과 속성값으로 구성된 입력값을 추가로 받을 수 있습니다.
plot(y,"LineWidth",5)

plot 함수에서는 플로팅하려는 인수와 선 사양을 지정하고 그 뒤에 속성 이름과 속성값을 추가로 지정합니다.
plot(x,y,"ro-","LineWidth",5)
