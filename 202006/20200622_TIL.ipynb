배열이 좀 신기하다.
x = [7, 9]
를 하면 
x = 
    7        9
이렇게 출력된다.

x = [7;9]
를 하면
x = 
    7
    9
이렇게 출력된다.

즉 행으로 만들 때는 ,를 사용하고 열을 만들 때는 ;를 사용한다.

x = [1 2 3;4 5 6]
하면
x = 1 2 3
    4 5 6
이렇게 출력된다.

수식도 가능
x = [sqrt(10) pi^2]

----------------------------------------------------

y = A(5, 7)
A의 5번째 행, 7번째 열의 값을 추출할 수 있다.

y = A(end, 2)
end를 행 인덱스 또는 열 인덱스로 사용하면 마지막 요소를 참조할 수 있다.

y = A(end-1, end-2)
이렇게도 사용 가능

콜론 연산자(:)를 인덱스로 사용하면 해당 차원의 모든 요소를 지정하게 된다.
x = A(2, :)
A의 두 번째 행에 있는 모든 요소를 포함하는 행 벡터를 만든다.
x = A(:, 2)
A의 두 번째 열에 있는 모든 요소를 포함하는 열 벡터를 만든다.

x = A(1:3, :)
A의 첫 번째 행, 두 번째 행 및 세 번째 행을 포함하는 행렬을 만든다.
volumes = data(:, end-1:end)
data의 마지막 두 개 열을 포함하는 변수

x = v(3)
v가 행 벡터 또는 열 벡터인 경우 벡터 v의 세 번째 요소를 반환한다.
x = v(3:end)
이런식으로도 사용가능

v2(1) = 0.5
v2의 첫 번째 요소를 0.5로 변경한다.

data(1, end) = 0.5
data의 첫 번째 행, 마지막 열의 요소 값을 0.5로 변경
