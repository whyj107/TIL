## 컴퓨터 시스템의 핵심 - OS란?
OS(Operating System) : 어떤 프로그램을 사용했을 때 그걸 작동시키는 환경

### 컴퓨터의 구성
CPU : 0과 1이라는 두가지 숫자로 이루어지는 신호들을 처리하는 장치. 
        프로그램의 명령어를 해석해서 0,1로 이루어진 연선 처리를 하고 제어하는 역할
        
메모리 : CPU가 명령을 처리할 때 옆에서 기억해주는 역할
         컴퓨터가 종료되면 메모리의 정보들은 저부 삭제된다(단기 기억용)
           
디스크 : 메모리와 달리 계속해서 저장해주는 장치
         영구적으로 저장되어야 할 것들을 보관(HDD, SDD)

OS의 역할 : 프로그램이 제대로 돌아가도록 CPU, 메모리에게 일을 하라는 명령을 내리는 역할(자원할당)

### OS에서 프로그램을 할당하는 방법, 프로세스
OS가 프로그래밍적으로 우선순위를 따져 후순위의 프로세스를 죽이는 관리를 하게 된다(자원 관리)
프로세스 : OS가 CPU와 메모리 자원을 활용해서 실행시킨 프로그램
           프로세스 안에는 Code, Data, Stack, Heap이 존재

### OS는 다 다르다
OS가 다르다는 건 아예 다른 프로그램이 사용되어야 한다.

### 추가설명
스레드 : 프로세스 내부에서 작동하게 되는 실행 단위
         OS에서 프로세스를 관리하며 프로그램을 실행시켰다면, 프로세스 안에선 스레드들이 실행되는 구조
    
커널 : OS의 핵심 부분으로 자원 관리, 메모리 제어 등을 담당. OS에 포함되어 있는 것

인터럽트 : 입출력 장치(키보드, 마우스 등)나 예외 사항들이 발생하면 OS는 CPU에게 먼저 처리하라고 명령
           원래 프로그램 중지 후 인터럽트로 들어온 명령을 처리

## 컴퓨터 간의 역할 - 네트워크란?
네트워크 : 컴퓨터들끼리 서로 연결이 되서 통신하는 것
             연결이 된 후 신호를 보냄

### 우리가 어떻게 네트워크를 통해 클래스 101을 이용할까
1. 브라우저 -> OS : 사이트 정보 요청
2. OS --와이파이 or 랜선 -->컴퓨터(클래스101 사이트 정보를 가졌음) : 연결 요구 신호를 보냄
3. OS <--복잡한 네트워크 길들(꼬불꼬불꼬불)--컴퓨터 : 문제가 없으면 연결을 승인
4. OS -> 컴퓨터 : 클래스 101 사이트 정보 요청

### 클라이언트와 서버
클라이언트 : 요청하게 되는 주체
서버 : 정보를 주는 주체

### 물리적 관점에서 네트워크
유선 랜카드 : 랜선을 꽂아서 인터넷 사용
랜선 : 광케이블(0, 1이라는 컴퓨터용 신호를 빛으로 변환해서 전송 후 다시 그 빛이 0, 1이라는 전기적 신호로 변환)을 사용
무선 랜카드 : 와이파이를 통해 인터넷 사용. 공기를 통해 전파를 쏘는 방식

### IP & DNS
IP : 자신의 네트워크 주소
DNS : Domain Name System. DNS서버에서 영문 주소를 네트워크 주소인 IP로 변경해 줌

### 프로토콜에 대해
프로토콜 : 혹은 통신 프로토콜. 네트워크의 규격

## 프로그래밍 언어와 라이브러리에 대해
### 프로그래밍 언어는 어떻게 쓰이는 걸까
프로그래밍언어 : 사람이 이해할 수 있는 언어인 영어로 되어있음
기계어 : OS가 이해할 수 있는 언어
저급언어 : 어셈블리어
고급언어 : 사람이 쉽게 칠 수 있는 프로그래밍 언어
컴파일 : 고급언어 -> 저급언어로 변환

### 컴파일 언어와 인터프리터 언어
컴파일 언어 : 작성한 프로그래밍 언어들을 몽땅 기계어로 컴파일 한 후에 실행 파일을 만듬
인터프리터 언어 : 실행하고 나서 코드 한 줄 씩 컴파일을 시키면서 실시간으로 작동

### 라이브러리가 뭘까?
라이브러리 : 다른 개발자들이 이런 일련의 과정을 미리 정리해 놓은 코드 뭉치

## 프론트엔드, 백엔드에 대해 들어보셨나요?
### 프론트엔드와 백엔드란?
프론트엔드 : 유저들이 서비스를 이용할 때 쓰게되는 웹, 앱 등을 의미
백엔드 : 정보를 주는 서버를 포괄적으로 의미. 서비스와 관련된 모든 정보들을 저장하고 있고 프론트엔드에게 제공해주는 역할

### 실제로 어떻게 프론트, 백엔드가 연결될까
클래스 101에서 마이페이즈를 볼 때
1. 처음에 프론트엔드에서 유저가 로그인을 요청을 하게되면 프론트엔드는 유저에 대한 정보를 받아오기 위해 서버(백엔드)에게 요청을 하게 된다.
2. 서버는 유저에게 맞는 정보들을 그 안에서 찾아서 다시 프론트엔드에게 돌려주게 된다.
3. 프론트엔드는 받은 정보들을 가지고 사용자가 볼 수 있게 텍스트와 이미지등으로 보여준다.
